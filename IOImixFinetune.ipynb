```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer, GPT2Config
from transformers import TextDataset, DataCollatorForLanguageModeling
from transformers import Trainer, TrainingArguments

model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

```

```python
config = GPT2Config()
```

```python
config.vocab_size = tokenizer.vocab_size
```

```python
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
```

```python
model.resize_token_embeddings(len(tokenizer))
```

```python
from datasets import load_dataset

dataset = load_dataset("roneneldan/TinyStories", split = "train[:10%]")
dataset1 = load_dataset("text", data_files = '/content/drive/MyDrive/IOIDatasetv2.0.txt')
```

```python
from datasets import concatenate_datasets
dataset2 = concatenate_datasets([dataset,dataset1["train"]])
```

```python
dataset1 = dataset1['train']
```

```python
dataset1[0]
```

```python
from datasets import Dataset
import pandas as pd
df1 = pd.DataFrame(dataset)
df2 = pd.DataFrame(dataset1)
df3 = pd.concat([df1,df2])
new_dataset = Dataset.from_pandas(df3)
new_dataset.remove_columns('__index_level_0__')
```

```python
dataset2
```

```python
def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        max_length=50
    )
```

```python
dataset2[26000]
```

```python
mydataset = dataset2.map(tokenize_function, batched=True,remove_columns = dataset2.column_names)
```

```python
mydataset
```

```python
mydataset[6000]
```

```python
data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer, mlm=False)
```

```python
training_args = TrainingArguments(
        output_dir='./results',
        num_train_epochs=3,
        per_device_train_batch_size=10,
        per_device_eval_batch_size=10,
        weight_decay = 0.1,
        learning_rate = 1e-5,
        save_total_limit = 5
    )
```

```python
trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=mydataset
    )
```

```python

trainer.train()
```

```python
trainer.save_model('/content/drive/MyDrive/IOI/97ts3ioi3/')
```
